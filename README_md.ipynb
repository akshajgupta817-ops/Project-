**QUICK SORT DEMONSTRATOR**

This project is an interactive Python app that visually demonstrates the Quick Sort Algorithm. Users enter a list of numbers, and the app sorts them using a recursive Quick Sort implementation. The interface is built using Gradio, making the algorithm easy to test and understand.

**1. Problem Breakdown & Computational Thinking**

**Decomposition**

I broke the problem into the following steps:
  1. Take user input as string

  2. Split the string into individual numbers

  3. Convert each number from text to an integer

  4. Apply the quick sort algorithm

  5. Display the sorted list back to the user

**Pattern Recognition**

Quick sort repeatedly chooses a pivot, divides the list into 3 groups,
recursively sorts the left and right groups. This pattern repeats itself for
every part of the list until everything is sorted.


**Abstraction**

To help the beginners understand it clearer, the user doesn't need to understand recursion, the app doesn't hide pivot selection, list partitioning, and recursive calls. The user also sees only the input become the output.

**Algorigthm Design**

The input is a comma seperated list of integers like (10,9,1,2,)

The Processing:

1. Convert input text to integer list

2. Run Quick Sort, select pivot, paritition values, recursively sort actions

3. Combine results

The output is a space seperated sorted list of numbers like (1 2 9 10)


**STEPS TO RUN THE APP**

Run Locally using code pip install gradio python app.py

Another option to run the app is to visit the HuggingFace app.

**FEATURES**

Some features are that it has a beginner friendly algorithm implementation, clean gradio interface, supports negative numbers, error checking for invalid input, simple and readable recursion and finally output formatted clearly for learning purposes.
